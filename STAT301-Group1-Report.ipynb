## 1. Introduction

### 1.1 Background

Insurance companies typically charge premiums for customers who are likely to pass away shortly after purchasing one of their policies. This is because they are undertaking a greater risk of paying the entire policy before the customer has invested more than the policy's worth. In order to stay afloat, life insurance companies would need to predict when a person dies, but unless one has a time machine, it is impossible to accurately predict when a person dies. Hence, life insurance companies depend on another scale, Life Expectancy.

Life Expectancy is defined as the average number of years of life a person who has attained a given age can expect to live. Life insurance companies use this as a benchmark scale to set the pricing of life insurance policies (Social Security, 2019). If there is a huge gap between your age and your life expectancy, there is a lower risk for the insurance company and the policy will be set at a cheaper price. Although Life Expectancy is a more feasible scale to use, there is still ambiguity in accurately calculating one's Life Expectancy.

A growing body of research suggests that a person's Life Expectancy may be affected by a number of essential external factors, including but not limited to alcohol consumption level (the rate of people in a country drinking alcohol or not), regional differences, developing or developed country, BMI(body mass index), population, schooling(the education level of the population). 

### 1.2 Question

As such, we were motivated to determine the factors that best predicts one's life expectancy. In other words, we are trying to address the question: "What are the variables that best predict one's life expectancy?". 

### 1.3 Dataset


To answer this question, we will be using the Life Expectancy Dataset from Kaggle. The dataset consists of 2938 rows of Life Expectancy data collected from the World Health Organization (WHO) from 2000 to 2015 across 193 countries. Each row has 20 explanatory variable which can be grouped into several broad categories; Immunization-related factors, Mortality factors, Economical factors and Social factors(Kumar., 2017).

Before we begin our analysis, we need to ensure that our model assumptions are valid. Particularly, we need to ensure that the input variables we will be using to create our linear model doesn't have high correlation with one another (multicollinearity). Multicollinearity inflates the estimates of the standard errors of the Least Squares (LS) estimators of the regression coefficients, and makes the results less trustworthy. Hence, we will check if we have this issue using the VIF as our scale.

```r
model <- lm(life_expectancy~., data = le)
model.vif = data.frame(vif = vif(model))
model.vif
```

The VIF result of covariates `infant_deaths` and `under_five_deaths` are significantly higher (> 180) than what is empirically taken as a threashold to determine a high correlation (usually greater than 10 indicates high correlation). Thus, we will remove `infant_deaths` and verify that the VIF of `model1` without the `infant_deaths` variable is acceptable. 

```r
model1 = lm(life_expectancy ~ .-infant_deaths, data = le)
model1.vif = data.frame(vif = vif(model1))
model1.vif
```

After removing the `infant_deaths` the VIF results seems to be more reasonable. None of the values are extremely high, which indicates a relatively low correlation. So, we can start creating our predictive model without including the `infant_deaths` as a potential input variable.   

### 2.4 Train-Test Split

To avoid potential information leakage from the test set, which will result in adding bias to our model, we will split the data into a training set and testing set with a 75:25 ratio. We will then perform variable selection using the training set. Also, we want to have approximately the same proportions of data for the factor variable `status` to ensure the two subgroups `Developing` and `Developed` are properly represented in both train and test set. 

```r
set.seed(9122022)
le_split = initial_split(le, prop = 0.75, strata = status)
le_train = training(le_split)
le_test  = testing(le_split)
```

### 2.5 Variable Selection 

In this step, we will perform a preliminary variable selection using the forward selection method, while we removed `infant_deaths` as previously discussed. The forward selection process begins with a null model. Then iteratively, at each step, the model includes the variable that improves the model the best as the best subset model. The RSS (Residual Sum of Squares) is used as the criterion during the process. The model with the lowest RSS value is qualified as the best. 

```r
names <- colnames(select(le, -life_expectancy, -infant_deaths))
names

le_forward_sel <- regsubsets(
  x =  life_expectancy ~ .-infant_deaths, nvmax = length(names),
  data = le_train,
  method = "forward"
)

le_forward_summary <- summary(le_forward_sel)
le_forward_summary
```

```r
le_forward_summary_df <- tibble(
    n_input_variables = 1:19,
    RSQ = le_forward_summary$rsq,
    RSS = le_forward_summary$rss,
    ADJ.R2 = le_forward_summary$adjr2,
    Cp = le_forward_summary$cp,
    BIC = le_forward_summary$bic)
le_forward_summary_df
```

```r
plot(le_forward_summary$cp, main = "Cp for forward selection",
  xlab = "Number of Input Variables", ylab = "Rsq", type = "b", pch = 19,
  col = "red", xaxp = c(1,19,18))
```

According to the above output, we decide to select the model with 11 variables since it is the elbow point such that adjusted $R^2$ value decreases only by small amount as more variables are included, and Cp and BIC values only decrease a little as more variables are included. We also consider that 11 variables model is already complex and adding more variables do not improve the model in the statistical sense, the tradeoff is not worth it. 

```r
# the 11 selected variables and their coefficients
coef(le_forward_sel,11) %>% round(3)
```

Using the 11 variables selected by forward selection, we will train it using `lm()` with the training dataset and call it `red_md`. We will use this model to predict values of the life expectancy in the test set. 

```r
red_md <- lm(life_expectancy ~ year + status + adult_mortality + alcohol +
             percentage_expenditure + bmi + under_five_deaths + diphtheria + 
             hiv_aids + income_composition_of_resources + schooling, 
          data = le_train)
summary(red_md)
```

### 2.6 Model Assumptions Analysis

After selecting the variables and creating our linear model, we would like to assess its predictive capabilities by using it to predict our testing set. Before doing that, however, we need to ensure that the model follows the model assumptions of a linear model such as heteroscedasticity and normality of the error terms. To do that, we will create a residual plot as well as a Diagnostic Q-Q plot.

```r
plot(red_md, 1, main = "Full Model")
```

The residual plot shows a slight clustering where the Fitted Values are around 70, but we previously saw that the mean life expectancy of our sample was around 70, which explains why a lot of our predictions have fitted values of 70. Aside from that, the plot shows a relatively random scattered cloud of points, suggesting there is no heteroscedasticity in our model. 

```r
plot(red_md, 2, main = "Full Model")
```

From this Q-Q plot, we can see that most of the standardized residuals lie on the 45 degree dotted line with slight variation, which is expected. This means that the standardized residuals match the theoretical quantile of the standard normal distribution, and would suggest that the error terms have a normal distribution. Therefore, the model assumptions are valid and we can use`red_md` as our final reduced model.

### 2.7 Model Evaluation

To assess how well our reduced model predicts life expectancy, we will perform prediction on the test dataset and calculate the coefficients of determination ($R^2$) to evaluate the model's accuracy. 

```r
predicted_values_reduced <- predict(red_md, le_test)

# compute coefficients of determination 
r_squared = cor(le_test["life_expectancy"], predicted_values_reduced) ^ 2
r_squared[1, 1]
```

We obtained a $R^2$ score of 0.831483411372635. We can interpret this as "approximately 83% of variability observed in `life_expectancy` in the test set can be explained by our model". 

To get a relative idea of how well our reduced model predicts, we will compare it with the full model (using all input variables) to see whether it improves on the simple model. We can create the full model and use it to predict the test dataset. We will then calculate the Root Mean Squared Error(RMSE) as the scale of predictive performance. 

```r
# full model 
full_md = lm(life_expectancy ~ ., data = le_train)
```

```r
models <- rbind(
  tibble(
  Model = "OLS Full Regression",
  R_MSE = rmse(
        le_test$life_expectancy,
        predict(full_md, le_test)
      )
    ),
  tibble(
    Model = "OLS Reduced Regression",
    R_MSE = rmse(
        le_test$life_expectancy,
        predict(red_md, le_test))
  )
)

models
```

As the full model has a lower RMSE than the reduced model, it would suggest that the full model performs better. However, this can be due to sampling variation. 

### 2.8 Shrinkage Methods 

Perhaps our reduced model would be better if we selected variables more smoothly. Shrinkage methods will allow us to sacrifice bias for a lower variance to gain prediction performance. Hence, we will explore LASSO and ridge regression as a potential model. Firstly, we need to turn our datasets into a matrix so that it can be used by our `cv.glmnet`function.

```r
le_X_train <- le_train %>% select(-life_expectancy)  %>% mutate(status = if_else(status == "Developed", 1, 0)) %>% as.matrix()
le_Y_train <- le_train %>% select(life_expectancy)  %>% as.matrix()

le_X_test <- le_test %>% select(-life_expectancy) %>% mutate(status = if_else(status == "Developed", 1, 0)) %>% as.matrix()
le_Y_test <- le_test %>% select(life_expectancy)  %>% as.matrix()
```

Now, we need to determine the optimal penalty parameter, lambda, to use to create our LASSO Regression model. We will do so by using a cross validation method over a variety of lambdas and calculate the Mean Squared Error of the model using the specific value of lambda.

```r
le_cv_lambda_LASSO <- cv.glmnet(
  x = le_X_train, y = le_Y_train,
  alpha = 1,
  lambda = exp(seq(-5, 10, 0.1))
)

plot(le_cv_lambda_LASSO, main = "Lambda selection by CV with LASSO\n\n")
```

We can then use the lambda that results in the smallest expected test MSE to create our LASSO regression model. 

```r
le_lambda_min_MSE_LASSO <- round(le_cv_lambda_LASSO$lambda.min, 4)

le_LASSO_min <- glmnet(
  x = le_X_train, y = le_Y_train,
  alpha = 1,
  lambda = le_lambda_min_MSE_LASSO
)
```

We can perform the same steps to create our ridge regression model. The only difference is that in the glmnet argument, we will be using `alpha = 0` to signify ridge regression.

```r
le_cv_lambda_ridge <- cv.glmnet(
  x = le_X_train, y = le_Y_train,
  alpha = 0,
  lambda = exp(seq(-5, 10, 0.1))
)

plot(le_cv_lambda_ridge, main = "MSE of Ridge estimated by CV for different lambdas\n\n")
```

```r
le_lambda_min_MSE_ridge <- round(le_cv_lambda_ridge$lambda.min, 4)
le_lambda_min_MSE_ridge

le_ridge_min <- glmnet(
  x = le_X_train, y = le_Y_train,
  alpha = 0,
  lambda = le_lambda_min_MSE_ridge
)
le_ridge_min
```

Now that we have the LASSO and ridge regression models, we can use them to predict the test dataset and calculate their respective RMSEs. We can then compare the improved models with the full and reduced models we checked in section 2.7.

```r
le_X_test_lasso <- le_test %>% select(-life_expectancy) %>% mutate(status = if_else(status == "Developed", 1, 0)) %>% as.matrix()

le_test_pred_LASSO_min <- predict(le_LASSO_min, newx = le_X_test_lasso)
le_test_pred_ridge_min <- predict(le_ridge_min,newx = le_X_test_lasso)
```

```r
models <- rbind(
  models,
  tibble(
    Model = "LASSO Regression with minimum MSE",
    R_MSE = rmse(
        le_test_pred_LASSO_min,
        le_test$life_expectancy
    )
  ),
  tibble(
    Model = "Ridge Regression with minimum MSE",
    R_MSE = rmse(
        le_test_pred_ridge_min,
        le_test$life_expectancy
    )
  )
  )
models
```
